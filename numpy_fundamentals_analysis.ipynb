## 📦 Array Creation

NumPy arrays (ndarrays) are the core data structure.

### Key Points:
- Arrays must contain elements of the same data type
- Lists are automatically converted if possible
- Arrays can be 1D, 2D, or higher dimensional

```python
# 1D Array
data1 = [1, 2, 3, 4, 5]
arr1 = np.array(data1)

print("1D Array:")
print(arr1)


# 2D Array
data2 = [range(1, 5), range(5, 9)]
arr2 = np.array(data2)

print("\n2D Array:")
print(arr2)


# Convert back to list
print("\nConverted back to list:")
print(arr2.tolist())
```

## 🔄 Reshaping and Axis Manipulation

Reshaping allows us to change the structure of an array without changing its data.

### Key Concepts:
- Total number of elements must remain constant
- Shape defines dimensions: (rows, columns)
- Axis defines direction:
  - Axis 0 → rows
  - Axis 1 → columns

Understanding axes is critical for data manipulation and machine learning workflows.

```python
# Create array
arr = np.arange(10, dtype=float)

print("Original Array:")
print(arr)

# Reshape into 2 rows and 5 columns
arr_2d = arr.reshape((2, 5))

print("\nReshaped (2x5):")
print(arr_2d)
print("Shape:", arr_2d.shape)

# Reshape into 5 rows and 2 columns
arr_reshaped = arr_2d.reshape((5, 2))

print("\nReshaped (5x2):")
print(arr_reshaped)
```

### 🔍 Insight:
- Reshaping does not change the data, only its structure
- Incorrect reshape (wrong size) will raise an error
- Always ensure total elements match:
  10 elements → (2×5) or (5×2), not (3×3)

## 📐 Understanding Axis (Very Important)

Axis represents the dimension along which operations are performed.

For a 2D array:
- Axis 0 → vertical direction (down rows)
- Axis 1 → horizontal direction (across columns)

Think:
👉 Axis 0 = "down"
👉 Axis 1 = "across"

```python
arr = np.arange(10).reshape(2, 5)

print("Array:\n", arr)

# Sum across rows (axis=1)
print("\nSum across rows (axis=1):")
print(arr.sum(axis=1))

# Sum across columns (axis=0)
print("\nSum across columns (axis=0):")
print(arr.sum(axis=0))
```

### 🔍 Insight:
- axis=0 → collapse rows → gives column-wise results
- axis=1 → collapse columns → gives row-wise results
- Axis confusion is one of the most common beginner mistakes

## ➕ Adding a New Axis

Sometimes we need to convert a 1D array into:
- Column vector
- Row vector

This is essential in machine learning and matrix operations.

```python
a = np.array([0, 1])

print("Original Array:")
print(a)
print("Shape:", a.shape)

# Convert to column vector
a_col = a[:, np.newaxis]

print("\nColumn Vector:")
print(a_col)
print("Shape:", a_col.shape)

# Alternative syntax
a_col_alt = a[:, None]

print("\nAlternative Column Vector:")
print(a_col_alt)
```

### 🔍 Insight:
- Original shape: (2,)
- Column vector: (2,1)
- Adding axis changes dimensionality → critical for matrix multiplication

## 🔁 Transpose

Transpose swaps rows and columns.

```python
print("Column Vector:\n", a_col)

print("\nTranspose (Row Vector):")
print(a_col.T)
```

### 🔍 Insight:
- Transpose converts:
  (n,1) → (1,n)

## 🔥 Flatten vs Ravel (Copy vs View)

Flattening converts multi-dimensional arrays into 1D arrays.

### Two Methods:
- flatten() → returns a COPY
- ravel() → returns a VIEW (if possible)

This distinction is critical for memory and performance.

```python
arr = np.arange(10).reshape(2, 5)

flat = arr.flatten()

# Modify flattened array
flat[0] = 99

print("Flattened Array:")
print(flat)

print("\nOriginal Array (unchanged):")
print(arr)
```

```python
arr = np.arange(10).reshape(2, 5)

ravel = arr.ravel()

# Modify ravel array
ravel[0] = 99

print("Ravel Array:")
print(ravel)

print("\nOriginal Array (modified):")
print(arr)
```

### 🔍 Insight:
- flatten() → creates new memory → safe but slower
- ravel() → shares memory → faster but risky

👉 If you modify ravel → original array changes  
👉 If you modify flatten → original array stays same  

This concept is frequently asked in interviews.

## 🧠 Memory Behavior (Advanced Concept)

NumPy tries to avoid copying data to improve performance.

- View → shares memory
- Copy → independent memory

Why this matters:
- Faster computations
- Lower memory usage
- But can cause unintended side effects

## 🧠 NumPy Memory Model (Row-Major Order)

NumPy stores arrays in **row-major order (C-style)**.

### What does this mean?

For a 2D array:
- Data is stored row by row
- Axis 0 → rows
- Axis 1 → columns

For a 3D array:
- Axis 0 → "planes" (depth)
- Axis 1 → rows
- Axis 2 → columns

### Why this matters:
- Faster computation
- Efficient memory access
- Important for reshaping & slicing

## 🧪 Demonstration of Row-Major Order

```python
import numpy as np

# Create a 2D array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Original Array:\n", arr)

# Flatten the array
flat = arr.flatten()

print("\nFlattened Array:", flat)
```

### 🔍 Insight:
Flattened output → [1, 2, 3, 4, 5, 6]

👉 This proves:
NumPy stores elements **row by row**

## 🔬 Check Memory Order Explicitly

```python
print("Is C-contiguous (row-major)?", arr.flags['C_CONTIGUOUS'])
print("Is F-contiguous (column-major)?", arr.flags['F_CONTIGUOUS'])
```

### 🔍 Output:
C_CONTIGUOUS → True  
F_CONTIGUOUS → False  

👉 Confirms NumPy uses row-major storage

```python
# Column-major flatten
flat_F = arr.flatten(order='F')

print("Row-major flatten:", arr.flatten())
print("Column-major flatten:", flat_F)
```

### 🔍 Difference:

Row-major → [1, 2, 3, 4, 5, 6]  
Column-major → [1, 4, 2, 5, 3, 6]

👉 This shows how memory traversal changes

## 🎯 Real-World Importance

- Impacts performance in large datasets
- Important for reshaping operations
- Critical in deep learning (NumPy, PyTorch, TensorFlow)

## ⚡ Vectorized Operations in NumPy

Vectorization refers to performing operations on entire arrays **without using explicit Python loops**.

### Why is this powerful?

- Faster execution (runs in optimized C code)
- Cleaner and more readable code
- Less error-prone

👉 This is one of the **biggest advantages of NumPy**

## 🧪 Basic Vectorized Computations

```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

# Multiply entire array
print(arr * 10)

# Square root
print(np.sqrt(arr))

# Power
print(arr ** 2)
```

### 🔍 Insight:
- Operations apply **element-wise**
- No loops needed
- Extremely fast compared to Python loops

## 🔢 Element-wise Array Operations

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
```

### 🔍 Insight:
Each element in `a` is operated with corresponding element in `b`

👉 This is NOT matrix multiplication — it's element-wise

## 📊 NumPy Universal Functions (ufuncs)

```python
arr = np.array([1, 4, 9, 16])

print("Square root:", np.sqrt(arr))
print("Exponential:", np.exp(arr))
print("Log:", np.log(arr))
```

### 🔍 Common ufuncs:
- np.sqrt()
- np.exp()
- np.log()
- np.sin(), np.cos()

👉 These are optimized and very fast

## 📏 Statistical Functions

```python
data = np.random.randn(4, 3)

print("Array:\n", data)

print("Mean:", data.mean())
print("Std Dev:", data.std())
print("Min:", data.min())
print("Max:", data.max())
print("Sum:", data.sum())
```

## 📊 Axis-Based Computation

```python
data = np.array([[1, 2, 3],
                 [4, 5, 6]])

# Column-wise (axis=0)
print("Sum (axis=0):", data.sum(axis=0))

# Row-wise (axis=1)
print("Sum (axis=1):", data.sum(axis=1))
```

- axis = 0 → operate DOWN (column-wise)

## 🔍 Boolean Vectorization

```python
arr = np.array([1, 5, 10, 15])

print(arr > 5)
print((arr > 5).sum())
print((arr > 5).any())
print((arr > 5).all())
```

### 🔍 Meaning:

- any() → at least one True

## 🔥 Broadcasting in NumPy

Broadcasting allows NumPy to perform operations on arrays of **different shapes**.

👉 Instead of copying data, NumPy **virtually stretches** smaller arrays.

### Why it matters:
- Eliminates loops
- Saves memory
- Enables fast computations

## 🧪 Example 1: Adding Scalar to Array

```python
import numpy as np

arr = np.array([1, 2, 3])

print(arr + 5)
```

### 🔍 Insight:
Scalar `5` is automatically applied to every element

👉 Internally: [5,5,5]

## 🧪 Example 2: Row + Column

```python
a = np.array([[0, 0, 0],
              [10, 10, 10],
              [20, 20, 20]])

b = np.array([1, 2, 3])

print(a + b)
```

### 🔍 What happens?

b = [1, 2, 3]

becomes:

[[1, 2, 3],
 [1, 2, 3],
 [1, 2, 3]]

👉 Then added row-wise

## 📏 Broadcasting Rules

NumPy compares shapes from **right to left**

Rule 1: Dimensions must be equal  
Rule 2: OR one of them must be 1  
Rule 3: Otherwise → ERROR

a = np.ones((3, 3))
b = np.array([1, 2, 3])

print(a + b)   # Works

```python
a = np.ones((3, 3))
b = np.array([1, 2, 3]) # Changed to 3 elements to match the last dimension of 'a'

print(a + b)   # Works now
```

Shapes (3,3) and (2,) → not compatible

👉 Broadcasting fails

## 🧪 Example 3: Column Broadcasting

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

mean = a.mean(axis=0)

print("Mean:", mean)
print("Centered Data:\n", a - mean)
```

### 🔍 What happens?

mean = [2.5, 3.5, 4.5]

is stretched to:

[[2.5, 3.5, 4.5],
 [2.5, 3.5, 4.5]]

👉 Then subtraction happens

## 🧪 Example 4: Normalization

```python
a = np.array([[1, 2, 3],
              [4, 5, 6]])

normalized = (a - a.mean(axis=0)) / a.std(axis=0)

print(normalized)
```

### 🔍 Used in:
- Machine Learning preprocessing
- Feature scaling
- Data normalization

## 🎯 Final Summary

👉 No loops

👉 Super fast
